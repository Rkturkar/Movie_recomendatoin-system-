# NLP Text processing 

Removing stop words(and , is , an, when,that,the) and puchuation (,-.--)

```python
df['tags'][1]
```

```python
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re
```

```python
nltk.download('stopwords')
nltk.download('wordnet')
```

```python
stop_words = set(stopwords.words('english'))
lemmatizer = WordNetLemmatizer()
```

words tokeniztion 

```python
def preprocess_text(text):
  #lowercase
  text = str(text).lower()
  #removed the punctuations
  text = re.sub(r'[^a-zA-Z\s]', '', text)

  words = text.split()
  #remove stopwords
  words = [word for word in words if word not in stop_words]

  #lemmatize
  words = [lemmatizer.lemmatize(word) for word in words]

  return " ".join(words)
```

```python
df['tags'] = df['tags'].apply(preprocess_text)
```

```python
from rapidfuzz import process
from sklearn.metrics.pairwise import cosine_similarity

def recommend(title, n=15):
    title = title.strip().lower()

    try:
        match, score, _ = process.extractOne(
            title,
            indices.index
        )

        # ❌ Totally wrong movie
        if score < 70:
            return ["Movie can't be found"]

        # ⚠️ Spelling mistake → suggest
        if 70 <= score < 90:
            return [f"Did you mean {match.title()}?"]

        # 🛡️ Extra safety: check index existence
        if match not in indices:
            return ["Movie can't be found"]

        idx = indices[match]

        # 🛡️ Bounds check
        if idx >= tfidf_matrix.shape[0]:
            return ["Movie can't be found"]

        sim_score = cosine_similarity(
            tfidf_matrix[idx],
            tfidf_matrix
        ).flatten()

        similar_idx = sim_score.argsort()[::-1][1:n+1]

        return df['title'].iloc[similar_idx].tolist()

    except Exception:
        return ["Movie can't be found"]
```

```python
recommend('Avtar')
```
